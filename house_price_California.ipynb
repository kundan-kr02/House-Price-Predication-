importing the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

importing the california house price dataset

```python
# house_price_dataset=sklearn.datasets.load_boston()
from sklearn.datasets import fetch_california_housing

# Load the California housing dataset
house_price_dataset = fetch_california_housing()

# Now you can access the data and target variables
X = house_price_dataset.data  # Features
y = house_price_dataset.target  # Target variable
```

```python
print(house_price_dataset)
```

```python
# loading the datasets to a panda dataset
house_price_dataframe=pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
# print the first five rows
house_price_dataframe.head()
```

```python
# add the target column (price) to the dataframe
house_price_dataframe['price']=house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# checking the number of rows and columns in the dataframe
house_price_dataframe.shape
```

```python
# check for missing values
house_price_dataframe.isnull().sum()
```

```python
# statistical measures of the dataset
house_price_dataframe.describe()
```

understanding the correlation between various features in the dataset

1. positive correlation
2. negative correlation

```python
correlation=house_price_dataframe.corr()
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

splitting the data and target

```python
X=house_price_dataframe.drop(['price'],axis=1)
Y=house_price_dataframe['price']
```

```python
 print(X)
 print(Y)
```

splitting the data into test data and training data

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

XGBoost Regressor

```python
# loading the model
model=XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train,Y_train)
```

Evaluation

prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction=model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1=metrics.r2_score(Y_train,training_data_prediction)

# mean absolute error
score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)


print("R squared error:",score_1)
print("maen absolute error:",score_2)
```

visualizing the actual prices and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("actual prices")
plt.ylabel("predicted prices")
plt.title("actual prices vs predicted prices")
plt.show()
```

prediction on test data

```python
# accuracy for prediction on test data
test_data_prediction=model.predict(X_test)
```

```python
# R squared error
score_1=metrics.r2_score(Y_test,test_data_prediction)

# mean absolute error
score_2=metrics.mean_absolute_error(Y_test,test_data_prediction)


print("R squared error:",score_1)
print("maen absolute error:",score_2)
```

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("actual prices")
plt.ylabel("predicted prices")
plt.title("actual prices vs predicted prices")
plt.show()
```
